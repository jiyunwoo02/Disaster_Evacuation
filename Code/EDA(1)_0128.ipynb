```python
#3. 결측치 처리
#3-1. 위도와 경도 결측치 보강
#'위도','경도' 컬럼 중 적어도 하나의 컬럼에서 NaN을 가지는 데이터행 추출
missing_loca = df[df[['위도', '경도']].isnull().any(axis=1)].copy()
#reset_index(drop=False)
#df에서의 index행을 컬럼으로 저장하고 missing_loca의 새 index 만들었음
#df의 index행은 missing_loca에서 결측 처리 후 df에 합칠 때 사용
missing_loca = missing_loca.reset_index(drop=False)
missing_loca
```

```python
#'위도', '경도' 결측치 처리 시 필요한 컬럼만 남기고 삭제
missing_loca = missing_loca[['index', '대피소명', '도로명주소', '주소', '위도', '경도', '대피소번호']]
missing_loca
```

```python
#missing_loca에서'위도', '경도' 있는 데이터 지우기
one = missing_loca[missing_loca['위도'].isnull() ^ missing_loca['경도'].isnull()]
one
```

```python
#'문암군부대앞산 고지대'의 '주소' 컬럼은 광범위 주소로 컬럼값 변환
one['주소'] = one['주소'].replace('강원도 고성군 죽왕면 문암진리 산1-8', '강원도 고성군 죽왕면 문암진리')
missing_loca['주소'] = missing_loca['주소'].replace('강원도 고성군 죽왕면 문암진리 산1-8', '강원도 고성군 죽왕면 문암진리')
df['주소'] = df['주소'].replace('강원도 고성군 죽왕면 문암진리 산1-8', '강원도 고성군 죽왕면 문암진리')
#데이터프레임 one의 '주소' 컬럼 활용해 위도 경도 결측치 채우기
one_add_list = one['주소'].tolist()
one_add_list
```

```python
#geopy 라이브러리 활용해 지오코딩으로 위도 경도 좌표 얻기
!pip install geopy
```

```python
from geopy.geocoders import Nominatim

def geocoding(address):
    geolocoder = Nominatim(user_agent = 'South Korea', timeout = None)
    geo = geolocoder.geocode(address)
    crd = {"lat": str(geo.latitude), "lng": str(geo.longitude)}

    return crd


for add in one_add_list:
    crd = geocoding(f"add")
    print(add)
    print(crd['lat'])
    print(crd['lng'])
```
